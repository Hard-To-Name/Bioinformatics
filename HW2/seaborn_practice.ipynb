```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f = open("mytestfile.fastq", 'r')
lines = f.readlines()
means = []
medians = []
mins = []
maxs = []

for i in range(0, len(lines), 4):
    dna = list(lines[i + 1].rstrip())
    quality = list(lines[i + 3].rstrip())
    prob = [1 - 10 ** ((33 - ord(q)) / 10) for q in quality]
    
    temp = pd.DataFrame({'dna': dna, 'prob': prob})
    means.append(temp.mean().values[0])
    medians.append(temp.median().values[0])
    mins.append(temp['prob'].min())
    maxs.append(temp['prob'].max())
    
    print("Mean for sequence %s: %s" % (i // 4 + 1, means[-1]));
    print("Median for sequence %s: %s" % (i // 4 + 1, medians[-1]));
    print("Min for sequence %s: %s" % (i // 4 + 1, mins[-1]));
    print("Max for sequence %s: %s" % (i // 4 + 1, maxs[-1]));
    print()
```

```python
fastq_data = pd.DataFrame({'mean': means, 'median': medians, 'min': mins, 'max': maxs})
f, axs = plt.subplots(1, 2)

# Plot 1: x axis - mean; y axis - median
sns.distplot(fastq_data['mean'], bins = 10, kde = False, ax = axs[0])
sns.scatterplot(fastq_data['mean'], fastq_data['median'], ax = axs[1])
plt.tight_layout()
```

```python
# Plot 2: x axis - median; y axis - min
f, axs = plt.subplots(1, 2)
sns.distplot(fastq_data['median'], bins = 10, kde = False, ax = axs[0])
sns.scatterplot(fastq_data['median'], fastq_data['min'], ax = axs[1])
plt.tight_layout()
```

```python
# Plot 3: x axis - median; y axis - max
f, axs = plt.subplots(1, 2)
sns.distplot(fastq_data['median'], bins = 10, kde = False, ax = axs[0])
sns.scatterplot(fastq_data['median'], fastq_data['max'], ax = axs[1])
plt.tight_layout()
```
